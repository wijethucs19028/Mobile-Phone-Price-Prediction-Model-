```python
# **** Define 4 categories for rear camera

categories = {
    'Dual Rear': ['dual rear'],
    'Triple Rear': ['triple rear'],
    'Quad Rear': ['quad rear'],
    'Other camera': []
}

# Function to categorize camera configurations
def categorizeCamera(camera_text):
    camera_text = camera_text.lower()
    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword in camera_text:
                return category
    return 'Other camera'

# Apply the categorization function to the 'Camera' column
data['Camera'] = data['Camera'].apply(categorizeCamera)

# Display the value counts of the new categories in the 'Camera' column
data['Camera'].value_counts()
```

```python
import re # to work with regular expressions in Python

#**** Define the categories based on screen width using regex to extract width
def categorize_resolution(resolution):
    match = re.search(r'(\d{2,4}) x (\d{2,4})', resolution)  #the function uses re.search to look for a pattern in the resolution string that matches (\d{2,4}) x (\d{2,4}).
    if match:
        width = int(match.group(1))
        if width <= 720:
            return 'HD (720p)'
        elif width <= 1080:
            return 'Full HD (1080p)'
        elif width <= 1440:
            return 'Quad HD (1440p)'
        else:
            return 'Other screen'
    else:
        return 'Unknown'

data['Screen_resolution']=data['Screen_resolution'].apply(categorize_resolution) 
```

```python
# Function to standardize core count descriptions
def standardize_processor(processor):
    processor = processor.lower()
    # Standardize core counts
    if 'octa core' in processor:
        return 'Octa Core'
    elif 'quad core' in processor:
        return 'Quad Core'
    elif 'deca core' in processor:
        return 'Deca Core'
    elif 'nine core' in processor or 'nine-cores' in processor or 'nine cores' in processor:
        return 'Nine Core'
    # If it specifies only clock speed, categorize as 'Other'
    elif re.search(r'\d+(\.\d+)?\s*ghz', processor):
        return 'Other processor'
    else:
        return 'Unknown'

data['Processor'] = data['Processor'].apply(standardize_processor)
```

```python
#***** Function to categorize external memory
def categorize_external_memory(external_memory):
    if "Not Supported" in external_memory:
        return "Not Supported"
    
    match = re.search(r'(\d+)\s?(GB|TB)', external_memory)
    capacity = 0
    if match:
        capacity = int(match.group(1))
        unit = match.group(2)
        if unit == 'TB':
            capacity *= 1024
    
    if "Memory Card Supported" in external_memory:
        if capacity == 0:
            return "Supported, Unknown Capacity"
        elif capacity <= 128:
            return "Supported, Low Capacity"
        elif capacity <= 512:
            return "Supported, Medium Capacity"
        elif capacity <= 2048:
            return "Supported, High Capacity"
    elif "Memory Card (Hybrid)" in external_memory:
        if capacity == 0:
            return "Hybrid, Unknown Capacity"
        elif capacity <= 512:
            return "Hybrid, Medium Capacity"
        elif capacity <= 2048:
            return "Hybrid, High Capacity"
    
    return "Unknown"    # If these unrelated entries (like android,camera...) are passed to the function, they will fall into the "Unknown" category.
```

```python
data['External_Memory'] = data['External_Memory'].apply(categorize_external_memory)
```

```python
import re

def categorize_fast_charging(speed):
    match = re.search(r'(\d+)', speed)   #regular expression r'(\d+)' is used to extract the numeric part of the string. 
    if match:
        speed_value = int(match.group(1))
        if speed_value >= 100:
            return 'Ultra Fast Charging'
        elif speed_value >= 50:
            return 'Super Fast Charging'
        elif speed_value >= 30:
            return 'Fast Charging'
        elif speed_value >= 20:
            return 'Moderate Charging'
    return 'Basic Charging' #Charging speeds below 20W.

# Apply the categorize_fast_charging function to the "fast_charging" column
data['fast_charging'] = data['fast_charging'].apply(categorize_fast_charging)
```

```python
from sklearn.model_selection import train_test_split # type: ignore
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.24,random_state = 4)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc=model.score(x_test,y_test)
    print(str(model)+'-->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression # type: ignore
lr=LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso # type: ignore
lasso=Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor # type: ignore
dt=DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor # type: ignore
rf=RandomForestRegressor()
model_acc(rf)
```

```python
#hyper parameter tunning
from sklearn.model_selection import GridSearchCV # type: ignore

parameters={'n_estimators':[10,50,100],'criterion':['squared_error','absolute_error','poisson']}
grid_obj=GridSearchCV(estimator=rf,param_grid=parameters)

grid_fit=grid_obj.fit(x_train,y_train)

best_model=grid_fit.best_estimator_
best_model



```
